#### Other ways to read json 

Besides using pandas to import JSON file, we can use python json library to import json data. Python treats json data as dictionary.

```python
import json 

with open('data/population_data.json') as f:
    json_data = json.load(f)

print(json_data[10])
print('\n')

print("Country Name: ", json_data[0]['Country Name'])
print("Country Code: ", json_data[0]['Country Code'])
print("\n")
```
